```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder, OrdinalEncoder
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

### Target Variable y is Outcome

```python
y = df['Outcome']
```

```python
X = df.drop('Outcome',axis = 1)
```

```python
print(y.shape),print(X.shape)
```

### Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.70,stratify = y, random_state = 40)
```

```python
print(X_train.shape,X_test.shape)
```

```python
print(y_train.shape,y_test.shape)
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler
import pickle
import pandas as pd

# Assume X_train is your training data (make sure it's a DataFrame or a 2D array)
scaler = StandardScaler()
df_transformed = scaler.fit_transform(X_train)  # Fit the scaler to training data and transform it

# Convert transformed data to a DataFrame if you want to view or save it
df_transformed = pd.DataFrame(df_transformed, columns=X_train.columns)  # Assuming X_train is a DataFrame
print(df_transformed)

# Save the scaler to a file
scaler_path = r'C:\Users\aresa\OneDrive\Desktop\Machine Learning\scaler.pkl'
with open(scaler_path, 'wb') as file:
    pickle.dump(scaler, file)

print("Scaler saved at:", scaler_path)
```

```python
model_x_train_transformed = pd.DataFrame(scaler.fit_transform(X_train),columns = X_train.columns, index = X_train.index)
```

```python
model_x_train_transformed
```

```python
model_x_test_transformed = pd.DataFrame(scaler.transform(X_test),columns = X_test.columns, index = X_test.index)
```

```python
model_x_test_transformed
```

## Normalization

```python
normal = MinMaxScaler()
```

```python
model_norm_x_train_transformed = pd.DataFrame(normal.fit_transform(X_train),columns = X_train.columns, index = X_train.index)
```

```python
model_norm_x_train_transformed
```

```python
model_norm_x_test_transformed = pd.DataFrame(normal.fit_transform(X_test),columns = X_test.columns, index = X_test.index)
```

```python
model_norm_x_test_transformed
```

### Training The Model

```python
model = SVC(kernel = 'linear')
```

```python
trained_model = model.fit(model_x_train_transformed,y_train)
```

### Model Evaluation

#### Accuracy Score of the Training Data

```python
X_pred = trained_model.predict(model_x_train_transformed)
accuracy = accuracy_score(X_pred,y_train)

print('Accuracy Score of the Training Data',accuracy)
```

#### Accuracy Score of the Test Data

```python
X_pred = trained_model.predict(model_x_test_transformed)
accuracy = accuracy_score(X_pred, y_test)

print('Accuracy Score of the Test Data',accuracy)
```

### Making the Predictive System 

```python
input_data = (2,197,70,45,543,30.5,0.158,53)

numpy_data = np.asarray(input_data)
#print(numpy_data)

reshaped_data = numpy_data.reshape(1,-1)
#print(reshaped_data)

prediction = scaler.transform(reshaped_data)
#print(prediction)

model_pred = model.predict(prediction)
print(model_pred)

if model_pred[0] == 0:
    print('The Person is Not Diabetic')
else:
    print('The Person is Diabetic ')
```
